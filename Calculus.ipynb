```python
def finding_intercept_equation_second_degree(a, b, c):

  y_intercept = (0, c)

  delta = b**2 - 4*a*c

  if delta > 0:
    x_intercept_1 = ((-b + math.sqrt(delta)) / (2*a), 0)
    x_intercept_2 = ((-b - math.sqrt(delta)) / (2*a), 0)
    print ("x-intercept:", x_intercept_1, ", y-intercept:", y_intercept)
    print ("x-intercept:", x_intercept_2, ", y-intercept:", y_intercept)
    return y_intercept, x_intercept_1, x_intercept_2

  elif delta == 0:
    x_intercept = ((-b / (2*a)), 0)
    print ("x-intercept:", x_intercept, ", y-intercept:", y_intercept)
    return y_intercept, x_intercept

  else:
    print("phương trình không giao với đường Ox,", "y-intercept:", y_intercept)
    return y_intercept
```

```python
# 20
print("câu 20")
finding_intercept_equation_second_degree(4, 0, 3)

print("-----------------------------------------------")
# 21.

print("câu 21")
finding_intercept_equation_second_degree(1, 1, -2)
```
